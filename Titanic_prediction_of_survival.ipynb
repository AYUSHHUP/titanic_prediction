```
gender=pd.get_dummies(data['Sex'],drop_first=True)
```

```
data['Gender']=gender
```

```
data
```

```
data.drop('Cabin',axis=1,inplace=True)
```

```
data
```

```
data.drop(['Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```
data
```

```
x=data[['PassengerId','Pclass','Age','SibSp','Parch','Fare','Gender']]
y=data['Survived']
```

```
y
```

```
#for prediction
from sklearn.model_selection import train_test_split
```

```
A_train, A_test, B_train, B_test = train_test_split(x,y, test_size=0.33, random_state=42)
```

```
from sklearn.linear_model import LogisticRegression
```

```
l=LogisticRegression()
```

```
l.fit(A_train,B_train)
```

```
predict=l.predict(A_test)
```

```
from sklearn.metrics import confusion_matrix
```

```
pd.DataFrame(confusion_matrix(B_test,predict),columns=['Predicted No','Predicted Yes'],index=['Actual No','Actual Yes'])
```
